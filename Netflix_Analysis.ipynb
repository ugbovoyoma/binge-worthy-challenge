```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import plotly.express as px
import datasist as ds
from collections import Counter
```

```python
data=pd.read_csv('/content/dataset.csv')
```

```python
# Calculate the counts of each type
type_counts = data['type'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
type_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#66c2a5', '#fc8d62'], explode=(0.1, 0), labels=type_counts.index)
plt.title('Proportion of Movies vs. TV Shows', fontsize=16)
plt.ylabel('')  # Remove default y-label
plt.show()
```

```python
# Calculate counts and percentages
type_counts = data['type'].value_counts()
total = type_counts.sum()
percentages = (type_counts / total) * 100

# Print the summary
print(f"Movies: {type_counts['Movie']} ({percentages['Movie']:.2f}%)")
print(f"TV Shows: {type_counts['TV Show']} ({percentages['TV Show']:.2f}%)")
```

```python
#top generes

# Split the 'listed_in' column by commas and flatten the list
all_genres = data['listed_in'].str.split(', ').sum()

# Count the occurrences of each genre
genre_counts = Counter(all_genres)

# Convert to a DataFrame for easier visualization
genre_df = pd.DataFrame(genre_counts.items(), columns=['Genre', 'Count']).sort_values(by='Count', ascending=False)

# Display top genres
print(genre_df.head(10))
```

```python
# Plot the top 10 genres
plt.figure(figsize=(12, 6))
sns.barplot(data=genre_df.head(10), x='Count', y='Genre', palette='viridis')
plt.title('Top 10 Most Common Genres on Netflix', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Genre', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python


# Convert 'date_added' to datetime format, handling errors
data['date_added'] = pd.to_datetime(data['date_added'], format='mixed', dayfirst=False, errors='coerce')


# Check for invalid dates converted to NaT
print("Invalid dates converted to NaT:", data['date_added'].isna().sum())

# Extract the year from 'date_added'
data['year_added'] = data['date_added'].dt.year

# Count additions by year
content_trends = data['year_added'].value_counts().sort_index()

# Display the trends
print(content_trends)
```

```python
# Plot the trends
plt.figure(figsize=(12, 6))
plt.plot(content_trends.index, content_trends.values, marker='o', linestyle='-', color='b')
plt.title('New Content Added to Netflix Over the Years', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Additions', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
# Group by year and type
type_trends = data.groupby(['year_added', 'type']).size().unstack()

# Plot trends by type
type_trends.plot(kind='line', figsize=(12, 6), marker='o')
plt.title('New Content Added to Netflix Over the Years by Type', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Additions', fontsize=12)
plt.legend(title='Type')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Separate Movies and TV Shows
movies = data[data['type'] == 'Movie']
tv_shows = data[data['type'] == 'TV Show']

# Extract numerical duration for Movies (strip " min")
movies['duration_minutes'] = movies['duration'].str.extract('(\d+)').astype(float)

# Extract numerical duration for TV Shows (strip " Season(s)")
tv_shows['seasons'] = tv_shows['duration'].str.extract('(\d+)').astype(float)

# Calculate averages
avg_movie_duration = movies['duration_minutes'].mean()
avg_tv_seasons = tv_shows['seasons'].mean()

# Display results
print(f"Average Movie Duration: {avg_movie_duration:.2f} minutes")
print(f"Average TV Show Duration: {avg_tv_seasons:.2f} seasons")
```

```python
import matplotlib.pyplot as plt

# Data for visualization
categories = ['Movies (Minutes)', 'TV Shows (Seasons)']
averages = [avg_movie_duration, avg_tv_seasons]

# Plot
plt.figure(figsize=(8, 5))
plt.bar(categories, averages, color=['#1f77b4', '#ff7f0e'])
plt.title('Average Duration: Movies vs TV Shows', fontsize=16)
plt.ylabel('Average Duration', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```
